```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets

# Load the digits dataset from scikit-learn
digits = datasets.load_digits()

# Preprocess the data
def preprocess_data(images, labels):
    # Flatten the images to 1D arrays
    images = images.reshape(images.shape[0], -1)

    # Add padding with size 1 to each side of the 1D arrays
    images_padded = np.pad(images, ((0, 0), (1, 1)), mode='constant', constant_values=0)

    return images_padded, labels

# Preprocess the data and split into train (60%), validation (20%), and test (20%) sets
x, y = preprocess_data(digits.images, digits.target)
x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size=0.4, stratify=y, random_state=42)
x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=0.5, stratify=y_temp, random_state=42)

# Display the shapes of the final datasets
print("Training set shape:", x_train.shape, y_train.shape)
print("Validation set shape:", x_val.shape, y_val.shape)
print("Test set shape:", x_test.shape, y_test.shape)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.model_selection import train_test_split

# Assume you have loaded and preprocessed the MNIST dataset (x_train, y_train, x_val, y_val, x_test, y_test)

def one_hot_encode(labels, num_classes=None):
    if num_classes is None:
        num_classes = np.max(labels) + 1
    return np.eye(num_classes)[labels]

def linear_activation(x):
    return x

def sigmoid_activation(x):
    return 1 / (1 + np.exp(-x))

def relu_activation(x):
    return np.maximum(0, x)

def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)

def cross_entropy_loss(predictions, labels):
    return -np.mean(labels * np.log(predictions + 1e-15))

def calculate_f1_score(y_true, y_pred):
    return f1_score(y_true, y_pred, average='weighted')

def train_model(x_train, y_train, x_val, y_val, activation_func, learning_rate=0.01, epochs=30):
    input_size = x_train.shape[1]
    output_size = len(np.unique(y_train))

    # Convert labels to one-hot encoding
    y_train_one_hot = one_hot_encode(y_train, num_classes=output_size)
    y_val_one_hot = one_hot_encode(y_val, num_classes=output_size)

    # Initialize weights and biases
    weights = np.random.randn(input_size, output_size)
    biases = np.zeros(output_size)

    # Store the learning curves
    train_losses = []
    val_losses = []

    for epoch in range(epochs):
        # Shuffle the training data
        permutation = np.random.permutation(len(x_train))
        x_train_shuffled = x_train[permutation]
        y_train_shuffled = y_train_one_hot[permutation]

        for i in range(len(x_train_shuffled)):
            inputs = x_train_shuffled[i]
            label = y_train_shuffled[i]

            # Forward pass
            logits = np.dot(inputs, weights) + biases
            predictions = softmax(logits)

            # Compute loss
            loss = cross_entropy_loss(predictions, label)

            # Backward pass (SGD update)
            error = predictions - label
            weights -= learning_rate * np.outer(inputs, error)
            biases -= learning_rate * error

        # Record training loss
        train_logits = np.dot(x_train, weights) + biases
        train_predictions = softmax(train_logits)
        train_loss = cross_entropy_loss(train_predictions, y_train_one_hot)
        train_losses.append(train_loss)

        # Record validation loss
        val_logits = np.dot(x_val, weights) + biases
        val_predictions = softmax(val_logits)
        val_loss = cross_entropy_loss(val_predictions, y_val_one_hot)
        val_losses.append(val_loss)

        print(f"Epoch {epoch + 1}/{epochs}, Training Loss: {train_loss:.4f}, Validation Loss: {val_loss:.4f}")

    return weights, biases, train_losses, val_losses

def evaluate_model(x_test, y_test, weights, biases):
    test_logits = np.dot(x_test, weights) + biases
    test_predictions = np.argmax(softmax(test_logits), axis=1)

    # F1 Score
    f1_score_result = calculate_f1_score(y_test, test_predictions)
    print(f"F1 Score: {f1_score_result:.4f}")

    # Confusion Matrix
    confusion_matrix_result = confusion_matrix(y_test, test_predictions)
    print("Confusion Matrix:")
    print(confusion_matrix_result)

    return f1_score_result, confusion_matrix_result

# Choose an activation function
activation_functions = [linear_activation, sigmoid_activation, relu_activation]

for activation_func in activation_functions:
    print(f"\nTraining model with {activation_func.__name__} activation:")
    weights, biases, train_losses, val_losses = train_model(x_train, y_train, x_val, y_val, activation_func)
    
    # Plot learning curves
    plt.plot(range(1, len(train_losses) + 1), train_losses, label='Training Loss')
    plt.plot(range(1, len(val_losses) + 1), val_losses, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Cross-Entropy Loss')
    plt.title(f'Model with {activation_func.__name__} Activation Learning Curves')
    plt.legend()
    plt.show()

    # Evaluate the model on the test set
    print(f"Evaluating model with {activation_func.__name__} activation on the test set:")
    f1_score_result, confusion_matrix_result = evaluate_model(x_test, y_test, weights, biases)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.model_selection import train_test_split


#Returns One-hot encoded array of labels
def one_hot_encode(labels, num_classes=None):
    if num_classes is None:
        num_classes = np.max(labels) + 1
    return np.eye(num_classes)[labels]

#Returns Output of the sigmoid activation applied to the input.
    
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
#Returns Output of the ReLU activation applied to the input.
def relu(x):
    return np.maximum(0, x)
#Returns Output of the softmax activation applied to the input.
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)
#Returns Cross-entropy loss between predictions and labels.
def cross_entropy_loss(predictions, labels):
    epsilon = 1e-15
    predictions = np.clip(predictions, epsilon, 1 - epsilon)
    return -np.mean(labels * np.log(predictions) + (1 - labels) * np.log(1 - predictions))
#Returns Dictionary containing weight and bias parameters for each layer.
def initialize_parameters(input_size, hidden_sizes, output_size):
    parameters = {}
    layer_sizes = [input_size] + hidden_sizes + [output_size]
    # Initialize weights and biases for each layer
    for i in range(1, len(layer_sizes)):
        parameters[f'W{i}'] = np.random.randn(layer_sizes[i-1], layer_sizes[i]) / np.sqrt(layer_sizes[i-1])
        parameters[f'b{i}'] = np.zeros((1, layer_sizes[i]))

    return parameters

def forward_propagation(x, parameters, activation_func):
    #Output of the neural network and a dictionary containing intermediate layer outputs.
    layer_outputs = {'A0': x}
    #Iterate through each layer for forward propagation
    for i in range(1, len(parameters)//2 + 1):
        W = parameters[f'W{i}']
        b = parameters[f'b{i}']
        Z = np.dot(layer_outputs[f'A{i-1}'], W) + b
        A = activation_func(Z)
        layer_outputs[f'Z{i}'] = Z
        layer_outputs[f'A{i}'] = A

    return layer_outputs[f'A{len(parameters)//2}'], layer_outputs

def backward_propagation(parameters, layer_outputs, x, y, activation_func_derivative, learning_rate, clip_value=1.0):
    #Returns Dictionary containing gradients for each parameter.
    m = x.shape[0]
    grads = {}
    #Compute gradient of the last layer
    dA = layer_outputs[f'A{len(parameters)//2}'] - y
    # Iterate through layers in reverse order for backward propagation
    for i in reversed(range(1, len(parameters)//2 + 1)):
        dZ = dA * activation_func_derivative(layer_outputs[f'Z{i}'])
        dW = np.dot(layer_outputs[f'A{i-1}'].T, dZ) / m
        db = np.sum(dZ, axis=0, keepdims=True) / m
        dA = np.dot(dZ, parameters[f'W{i}'].T)

        # Gradient clipping
        for grad in [dW, db]:
            np.clip(grad, -clip_value, clip_value, out=grad)

        parameters[f'W{i}'] -= learning_rate * dW
        parameters[f'b{i}'] -= learning_rate * db

    return parameters

def train_neural_network(x_train, y_train, x_val, y_val, hidden_sizes, activation_func, activation_func_derivative, learning_rate=0.01, epochs=30):
    input_size = x_train.shape[1]
    output_size = len(np.unique(y_train))

    # Convert labels to one-hot encoding
    y_train_one_hot = one_hot_encode(y_train, num_classes=output_size)
    y_val_one_hot = one_hot_encode(y_val, num_classes=output_size)

    # Initialize parameters
    parameters = initialize_parameters(input_size, hidden_sizes, output_size)

    # Store the learning curves
    train_losses = []
    val_losses = []

    for epoch in range(epochs):
        # Forward and backward propagation for each training example
        for i in range(len(x_train)):
            x = x_train[i:i+1]
            y = y_train_one_hot[i:i+1]

            # Forward propagation
            predictions, layer_outputs = forward_propagation(x, parameters, activation_func)

            # Compute loss
            loss = cross_entropy_loss(predictions, y)

            if np.isnan(loss):
                print(f"Epoch {epoch + 1}/{epochs}, Training Loss: NaN, Validation Loss: NaN")
                print("Numerical instability. Try reducing learning rate.")
                break

            # Backward propagation
            parameters = backward_propagation(parameters, layer_outputs, x, y, activation_func_derivative, learning_rate)

        # Record training loss
        train_predictions, _ = forward_propagation(x_train, parameters, activation_func)
        train_loss = cross_entropy_loss(train_predictions, y_train_one_hot)
        train_losses.append(train_loss)

        # Record validation loss
        val_predictions, _ = forward_propagation(x_val, parameters, activation_func)
        val_loss = cross_entropy_loss(val_predictions, y_val_one_hot)
        val_losses.append(val_loss)

        print(f"Epoch {epoch + 1}/{epochs}, Training Loss: {train_loss:.4f}, Validation Loss: {val_loss:.4f}")

    return parameters, train_losses, val_losses

def evaluate_neural_network(x_test, y_test, parameters, activation_func):
    test_predictions, _ = forward_propagation(x_test, parameters, activation_func)

    # F1 Score
    test_predictions_classes = np.argmax(test_predictions, axis=1)
    f1 = f1_score(y_test, test_predictions_classes, average='weighted')
    print(f"F1 Score: {f1:.4f}")

    # Confusion Matrix
    cm = confusion_matrix(y_test, test_predictions_classes)
    print("Confusion Matrix:")
    print(cm)

    return f1, cm

# Choose an activation function
activation_functions = {'linear': {'func': lambda x: x, 'derivative': lambda x: 1},
                        'sigmoid': {'func': sigmoid, 'derivative': lambda x: sigmoid(x) * (1 - sigmoid(x))},
                        'relu': {'func': relu, 'derivative': lambda x: (x > 0).astype(float)}}

hidden_sizes = [128, 64, 32]  # Adjust the number of hidden units as needed

for activation_func_name, activation_func in activation_functions.items():
    print(f"\nTraining neural network with {activation_func_name} activation:")
    parameters, train_losses, val_losses = train_neural_network(x_train, y_train, x_val, y_val, hidden_sizes,
                                                                activation_func['func'], activation_func['derivative'])

    # Plot learning curves
    plt.plot(range(1, len(train_losses) + 1), train_losses, label='Training Loss')
    plt.plot(range(1, len(val_losses) + 1), val_losses, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Cross-Entropy Loss')
    plt.title(f'Neural Network with {activation_func_name.capitalize()} Activation Learning Curves')
    plt.legend()
    plt.show()

    # Evaluate the model on the test set
    print(f"Evaluating neural network with {activation_func_name} activation on the test set:")
    f1_score_result, confusion_matrix_result = evaluate_neural_network(x_test, y_test, parameters, activation_func['func'])
```
